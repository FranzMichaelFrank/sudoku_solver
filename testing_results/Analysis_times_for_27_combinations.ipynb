## Time efficiency by indvidual methods

```python

se = times.groupby("selection").mean()
se["time_normalized"] = se["time"] / se["time"].min()
se
```

```python
#note: all the time values correspond to the time in seconds for 75 cycles
co = times.groupby("crossover").mean()
co["time_normalized"] = co["time"] / co["time"].min()
co
```

```python
#note: all the time values correspond to the time in seconds for 75 cycles
mu = times.groupby("mutation").mean()
mu["time_normalized"] = mu["time"] / mu["time"].min()
mu
```
