```python
import os, shutil
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
directory = "/Users/franz/Downloads/sudoku_solver-main-4/testing_results/2_fitness_functions/"

all_file_names = sorted(os.listdir(directory))
try:
    all_file_names.remove(".DS_Store")
except:
    pass

print("There are", len(all_file_names), "csv files in the directory!")
```

```python
tournament__pmx_co__swap_mutation__243__20000__co_p__0_7__mu_p__0_6 = pd.read_csv(directory + "tournament__pmx_co__swap_mutation__243__20000__co_p__0_7__mu_p__0_6.csv")
tournament__pmx_co__swap_mutation__270__20000__co_p__0_7__mu_p__0_6 = pd.read_csv(directory + "tournament__pmx_co__swap_mutation__270__20000__co_p__0_7__mu_p__0_6.csv")
```

```python
display(tournament__pmx_co__swap_mutation__243__20000__co_p__0_7__mu_p__0_6)
print("\n\n\033[1mFitness Function with 243 max:\033[0m")
print("Average fitness score throughout all the difficulty levels:", np.round(tournament__pmx_co__swap_mutation__243__20000__co_p__0_7__mu_p__0_6["Average_Fitness"].mean(),4))
print("Average percentage of 243 reached throughout all the difficulty levels:", np.round(tournament__pmx_co__swap_mutation__243__20000__co_p__0_7__mu_p__0_6["Average_Fitness"].mean()/243,4))

for i in range(3):
    count = 0
    for element in tournament__pmx_co__swap_mutation__243__20000__co_p__0_7__mu_p__0_6.iloc[i,:].tolist():
        if element == 243:
            count += 1
            
    if i == 0:
        difficulty = "easy"
    elif i == 1:
        difficulty = "medium"
    else:
        difficulty = "hard"
        
    print("Completely solved Sudokus for difficulty level", difficulty + ":", count)
```
